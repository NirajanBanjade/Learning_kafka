```python
from elasticsearch import Elasticsearch, helpers
import json
import certifi  # provides Mozilla's CA bundle

client = Elasticsearch(
    "https://yelp-deployment-86e590.es.eu-central-1.aws.cloud.es.io",
    verify_certs=True,                   # ✅ verify certs
    ca_certs=certifi.where(),            # ✅ use trusted CA bundle
    basic_auth=("elastic", "es7oUdRdTEGYIJI5DMli3cIn")
)

# Open dataset
with open("/kaggle/input/yelp-dataset/yelp_academic_dataset_business.json") as f:
    data = []
    for line in f:
        try:
            doc = json.loads(line.strip())
            data.append(doc)
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON: {e}")

    helpers.bulk(client, data, index="businesses")
```
